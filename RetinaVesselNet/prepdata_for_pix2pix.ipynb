```python
import os

%matplotlib inline
outputFolder='C:/Users/zecharpy/Desktop/RetinaVesselNet/pix2pixdata/train/'
max_imgs=100
img_path=train_imgs
mask_path=train_groudtruth
def convert_pix2pixdata(img_path,mask_path, outputFolder,max_imgs):
    counter=0
    img_ls=os.listdir(img_path)
    mask_ls=os.listdir(mask_path)
    # loop over the input house paths
    for im,mask in zip(img_ls, mask_ls):
        outputImage = np.zeros((128, 256, 3), dtype="uint8")
        print(im,mask)
        # update the list of input images
        image =cv2.imread(img_path+im)
        #mask=cv2.imread(mask_path+mask)
        mask=plt.imread(mask_path+mask)
        mask=cv2.cvtColor(mask, cv2.COLOR_GRAY2RGB)
        print(image.shape, mask.shape)
        if (type(image) is np.ndarray) and (type(mask) is np.ndarray):
            if counter < max_imgs:
                name=str(im.split('_')[0])
                print("image number",name)
                image = cv2.resize(image, (128, 128))
                mask = cv2.resize(mask,(128,128))
                outputImage[0:128, 0:128] = image
                outputImage[0:128, 128:256] = mask
                plt.imshow(outputImage[:,:,[2,1,0]])
                plt.title(name)
                plt.show()
                # write image to pix2pixdata folder
                cv2.imwrite(outputFolder+'{}.jpg'.format(name), outputImage) 
                counter +=1
            else:
                break

```

```python
convert_pix2pixdata(img_path,mask_path, outputFolder,max_imgs)
```
